```python
import numpy as np


arr1 = np.random.randint(1, 10, size=(5, 5))
arr2 = np.random.randint(1, 10, size=(5, 5))


addition = arr1 + arr2


subtraction = arr1 - arr2


multiplication = arr1 * arr2


division = np.divide(arr1, arr2)


print("arr1:")
print(arr1)
print("\narr2:")
print(arr2)


print("\nElement-wise Addition:")
print(addition)
print("\nElement-wise Subtraction:")
print(subtraction)
print("\nElement-wise Multiplication:")
print(multiplication)
print("\nElement-wise Division:")
print(division)
```

```python
import pandas as pd
import numpy as np


data = {
    'Name': ['Sina', 'Amir', 'Saeed', 'Soroush', 'Hesam'],
    'Score': [85, 92, 76, 68, 95]
}

df = pd.DataFrame(data)


conditions = [
    df['Score'] >= 90,
    df['Score'] >= 80,
    df['Score'] >= 70,
    df['Score'] < 70
]

grades = ['A', 'B', 'C', 'D']


df['Grade'] = np.select(conditions, grades, default='Unknown')


print(df)
```
