```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression, SGDClassifier
from sklearn.datasets import make_classification
```

```python
X, y= make_classification(n_samples=1000,
                          n_features=2,
                          n_informative=2,
                          n_redundant=0,
                          n_repeated=0,
                          n_classes=2,
                          n_clusters_per_class=1,
                          weights=None,
                          flip_y=0.0001,
                          class_sep=0.45,
                          hypercube=True,
                          shift=0.0,
                          scale=1.0,
                          shuffle=True,
                          random_state=73)
colors = np.array(['blue','red'])
plt.scatter(X[:,0],X[:,1], c=colors[y])
plt.show
```

```python
x_train, x_test, y_train, y_test = train_test_split(X,y, test_size=0.2)
```

```python
model = LogisticRegression(random_state=73)
model.fit(x_train,y_train)
```

```python
print("Number of iterations:", model.n_iter_)
print("Score for train:", model.score(x_train,y_train))
print("Score for test:", model.score(x_test,y_test))
```

```python
x1_min, x2_min = x_test.min(0)
x1_max, x2_max = x_test.max(0)
n=500
x1r = np.linspace(x1_min, x1_max, n)
x2r = np.linspace(x2_min, x2_max, n)
x1m , x2m = np.meshgrid(x1r, x2r)
xm = np.stack((x1m.flatten(), x2m.flatten()), axis=1)
ym = model.decision_function(xm)
colors = np.array(['blue','red'])
plt.scatter(x_test[:,0],x_test[:,1], c=colors[y_test])
plt.contour(x1m, x2m, ym.reshape(x1m.shape), levels=[-1,0,1])
plt.show
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(x_test, y_test, clf=model)

y_pred = model.predict(x_test)
misclassified_indices = (y_pred != y_test)
misclassified_points = x_test[misclassified_indices]

plt.scatter(misclassified_points[:, 0], misclassified_points[:, 1], marker='o', edgecolor='black', facecolor='red', s=50, label='Misclassified')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')

plt.legend()

plt.show()
```
